#### Problem 2: 

Create a vertical violin plot with the 'titanic' dataset to visualize the distribution of fare cost for different classes.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_dataset = sns.load_dataset('titanic')
df = sns.load_dataset('titanic')

sns.violinplot( x = 'class',
            y = 'fare',
            data = df,
            color = "Yellow")
 
# Show the plot
plt.show()
```
